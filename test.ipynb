```python
import pandas as pd

data = pd.read_csv('./aceh_production_data_daily_ed.csv')
data['WELL_BORE_CODE'].unique().tolist()
```

```python
def data_multiselect(values):
    return [{"label": value, "value": value} for value in values]
```

```python
data_multiselect(data['WELL_BORE_CODE'].unique().tolist())
```

```python
data['WATER_CUT_DAILY'] = (data['BORE_WAT_VOL'] / (data['BORE_OIL_VOL'] + data['BORE_GAS_VOL'] + data['BORE_WAT_VOL'])) * 100
data['GAS_OIL_RATIO'] = data['BORE_GAS_VOL'] / data['BORE_OIL_VOL']
```

```python
pivot = data.pivot_table(values= ['WATER_CUT_DAILY','BORE_OIL_VOL'],
                 index=[data['DATEPRD']],
                 aggfunc='mean',
                 dropna=True
                 )
```

```python
filtered = []
for index, row in pivot.iterrows():
    filtered.append({'DATEPRD': index, **row.to_dict()})
```
